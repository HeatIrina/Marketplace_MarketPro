```python
# Создаем словарь dataframes из существующих переменных
dataframes = {
    'categories': categories,
    'order_items': order_items,
    'orders': orders,
    'products': products,
    'reviews': reviews,
    'transactions': transactions,
    'users': users
}

# Проверяем, что все переменные определены
print("📊 ЗАГРУЖЕННЫЕ ТАБЛИЦЫ:")
print("-" * 40)

for name, df in dataframes.items():
    if df is not None:
        print(f"✅ {name.upper()}: {df.shape[0]} строк, {df.shape[1]} столбцов")
    else:
        print(f"❌ {name.upper()}: не загружена")

# Теперь выполняем проверку на дубликаты
print("\n" + "=" * 70)
print("ЭТАП 2: ПОДГОТОВКА ДАННЫХ")
print("=" * 70)

print("\n1. ПРОВЕРКА НА ДУБЛИКАТЫ")
print("-" * 40)

for table_name, df in dataframes.items():
    if df is not None:
        initial_rows = df.shape[0]
        duplicates = df.duplicated().sum()
        
        if duplicates > 0:
            print(f"📊 {table_name.upper()}:")
            print(f"   Начальное количество строк: {initial_rows}")
            print(f"   Найдено дубликатов: {duplicates} ({duplicates/initial_rows*100:.2f}%)")
            
            # Удаляем дубликаты
            dataframes[table_name] = df.drop_duplicates()
            
            # Проверяем результат
            final_rows = dataframes[table_name].shape[0]
            print(f"   Осталось строк после удаления: {final_rows}")
            print(f"   Удалено строк: {initial_rows - final_rows}")
            print(f"   ✓ Дубликаты удалены")
        else:
            print(f"📊 {table_name.upper()}: дубликатов не найдено")
    else:
        print(f"📊 {table_name.upper()}: таблица не загружена")
```

```python
# 2. ПРЕОБРАЗОВАНИЕ СТОЛБЦОВ С ДАТАМИ В DATETIME
print("\n\n2. ПРЕОБРАЗОВАНИЕ ДАТ В DATETIME")
print("-" * 40)
# Словарь: таблица -> список столбцов с датами
date_columns_map = {
    'users': ['registration_date'],
    'orders': ['order_date'],
    'reviews': ['review_date'],
    'transactions': ['transaction_date']
}

for table_name, date_columns in date_columns_map.items():
    df = dataframes.get(table_name)
    if df is not None:
        print(f"\n{table_name.upper()}:")
        for col in date_columns:
            if col in df.columns:
                original_type = df[col].dtype
                df[col] = pd.to_datetime(df[col])
                new_type = df[col].dtype
                print(f"  {col}: {original_type} → {new_type} ✓")
            else:
                print(f"  ⚠️  {col}: столбец не найден")
```

```python
# 3. ИСПРАВЛЕНИЕ ЛОГИЧЕСКИХ ОШИБОК (ОТРИЦАТЕЛЬНЫЕ ЗНАЧЕНИЯ ЦЕН)
print("\n\n3. ИСПРАВЛЕНИЕ ЛОГИЧЕСКИХ ОШИБОК")
print("-" * 40)

# Проверяем цены в таблице products
if 'products' in dataframes:
    products_df = dataframes['products']
    
    if 'price' in products_df.columns:
        # Находим отрицательные цены
        negative_prices = products_df[products_df['price'] < 0]
        negative_count = len(negative_prices)
        
        print(f"📊 PRODUCTS (цены товаров):")
        print(f"   Всего товаров: {len(products_df)}")
        print(f"   Товаров с отрицательной ценой: {negative_count}")
        
        if negative_count > 0:
            print(f"   Примеры товаров с отрицательной ценой:")
            print(negative_prices[['product_id', 'title', 'price']].head())
            
            # Заменяем отрицательные цены на 0
            dataframes['products'].loc[products_df['price'] < 0, 'price'] = 0
            
            # Проверяем результат
            fixed_negative = (dataframes['products']['price'] < 0).sum()
            print(f"   ✓ Отрицательные цены заменены на 0")
            print(f"   Осталось товаров с отрицательной ценой: {fixed_negative}")
        else:
            print(f"   ✓ Отрицательных цен не обнаружено")
```

```python
# 4. ДОБАВЛЕНИЕ КАТЕГОРИАЛЬНЫХ ПРИЗНАКОВ (КАТЕГОРИЗАЦИЯ РЕЙТИНГА)
print("\n\n4. ДОБАВЛЕНИЕ КАТЕГОРИАЛЬНЫХ ПРИЗНАКОВ")
print("-" * 40)

if 'reviews' in dataframes:
    reviews_df = dataframes['reviews']
    
    if 'rating' in reviews_df.columns:
        # Функция для категоризации рейтинга
        def categorize_rating(rating):
            if pd.isna(rating):
                return 'No Rating'
            elif rating >= 4.5:
                return 'High (4.5-5.0)'
            elif rating >= 4.0:
                return 'Medium (4.0-4.49)'
            else:
                return 'Low (< 4.0)'
        
        # Добавляем новый столбец
        dataframes['reviews']['rating_category'] = reviews_df['rating'].apply(categorize_rating)
        
        print(f"📊 REVIEWS (отзывы):")
        print(f"   Всего отзывов: {len(reviews_df)}")
        print(f"   Добавлен новый столбец: 'rating_category'")
        
        # Статистика по категориям
        rating_stats = dataframes['reviews']['rating_category'].value_counts().sort_index()
        print(f"   Распределение по категориям:")
        for category, count in rating_stats.items():
            percentage = (count / len(reviews_df)) * 100
            print(f"     {category}: {count} ({percentage:.1f}%)")
        
        print(f"   ✓ Категоризация рейтинга выполнена")
```

```python
# 5. АНАЛИЗ ПРОПУЩЕННЫХ ЗНАЧЕНИЙ
print("\n\n5. АНАЛИЗ ПРОПУЩЕННЫХ ЗНАЧЕНИЙ (NaN)")
print("-" * 40)

# Создаем сводную таблицу с информацией о пропусках
missing_summary = []

for table_name, df in dataframes.items():
    total_rows = df.shape[0]
    total_cells = df.size
    
    # Считаем пропуски по столбцам
    missing_by_column = df.isnull().sum()
    total_missing = missing_by_column.sum()
    missing_percentage = (total_missing / total_cells) * 100
    
    # Добавляем в сводную таблицу
    missing_summary.append({
        'Таблица': table_name.upper(),
        'Всего строк': total_rows,
        'Всего столбцов': df.shape[1],
        'Всего пропусков': total_missing,
        '% пропусков': f"{missing_percentage:.2f}%"
    })
    
    # Выводим детальную информацию о пропусках
    if total_missing > 0:
        print(f"\n📊 {table_name.upper()}:")
        print(f"   Всего пропусков: {total_missing} ({missing_percentage:.2f}% от всех ячеек)")
        print(f"   Столбцы с пропусками:")
        
        for column, missing_count in missing_by_column.items():
            if missing_count > 0:
                column_percentage = (missing_count / total_rows) * 100
                print(f"     {column}: {missing_count} пропусков ({column_percentage:.2f}% строк)")
    else:
        print(f"📊 {table_name.upper()}: пропусков нет ✓")

# Выводим сводную таблицу
print("\n" + "=" * 70)
print("СВОДНЫЙ ОТЧЕТ О ПРОПУЩЕННЫХ ЗНАЧЕНИЯХ")
print("=" * 70)

summary_df = pd.DataFrame(missing_summary)
print(summary_df.to_string(index=False))
```

```python
# 3. BOXPLOT ДЛЯ ОБНАРУЖЕНИЯ ВЫБРОСОВ
print("3. BOXPLOT ДЛЯ ОБНАРУЖЕНИЯ ВЫБРОСОВ")
print("-" * 70)
    
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
# Функция для анализа выбросов
def analyze_outliers(data, column_name, ax):
    """Анализирует и визуализирует выбросы"""
    clean_data = data.dropna()
        
# Вычисляем границы для выбросов
    Q1 = clean_data.quantile(0.25)
    Q3 = clean_data.quantile(0.75)
    IQR = Q3 - Q1
        
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
        
# Находим выбросы
    outliers = clean_data[(clean_data < lower_bound) | (clean_data > upper_bound)]
        
# Создаем boxplot
    box = ax.boxplot(clean_data, patch_artist=True)
    box['boxes'][0].set_facecolor('lightblue')
    box['medians'][0].set_color('red')
    box['medians'][0].set_linewidth(2)
        
    ax.set_title(f'Boxplot: {column_name}', fontsize=14)
    ax.set_ylabel('Значение', fontsize=12)
    ax.grid(True, alpha=0.3)
        
# Добавляем аннотации
    stats_text = f"Выбросов: {len(outliers)} ({len(outliers)/len(clean_data)*100:.1f}%)"
    if len(outliers) > 0:
        stats_text += f"\nMin выброс: {outliers.min():.2f}\nMax выброс: {outliers.max():.2f}"
        
        ax.text(0.02, 0.98, stats_text, transform=ax.transAxes,
               verticalalignment='top', fontsize=10,
               bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))
        
        return outliers
    
# Boxplot для price
outliers_price = analyze_outliers(products_analysis['price'], 'price', axes[0])
    
# Boxplot для stock_quantity
outliers_stock = analyze_outliers(products_analysis['stock_quantity'], 'stock_quantity', axes[1])
    
plt.tight_layout()
plt.show()

# Автоматический анализ выбросов
print("\n📊 АВТОМАТИЧЕСКИЙ АНАЛИЗ ВЫБРОСОВ:")

for column in ['price', 'stock_quantity']:
    data = products_analysis[column].dropna()
    
    # Вычисляем границы выбросов
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Находим выбросы
    outliers = data[(data < lower_bound) | (data > upper_bound)]
    
    print(f"\n🔹 {column.upper()}:")
    print(f"   Всего значений: {len(data)}")
    print(f"   Выбросов: {len(outliers)} ({len(outliers)/len(data)*100:.1f}%)")
    
    if len(outliers) > 0:
        print(f"   Минимальный выброс: {outliers.min():.2f}")
        print(f"   Максимальный выброс: {outliers.max():.2f}")
        print(f"   Среднее выбросов: {outliers.mean():.2f}")
    else:
        print(f"   ✓ Выбросов не обнаружено")
```

```python
# 4. КОРРЕЛЯЦИОННЫЙ АНАЛИЗ
print("4. КОРРЕЛЯЦИОННЫЙ АНАЛИЗ")
print("-" * 70)
    
# Вычисляем корреляционную матрицу
correlation_matrix = products_analysis.corr()
    
print("\n📊 КОРРЕЛЯЦИОННАЯ МАТРИЦА:")
print(correlation_matrix.round(3))
    
# Визуализируем корреляционную матрицу
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
# Тепловая карта корреляций
sns.heatmap(correlation_matrix, 
            annot=True, 
            fmt=".3f",
            cmap="coolwarm",
            center=0,
            square=True,
            cbar_kws={"shrink": 0.8},
            ax=axes[0])
axes[0].set_title('Тепловая карта корреляционной матрицы', fontsize=14)
    
# Scatter plot с линией регрессии
sns.regplot(data=products_analysis, 
            x='price', 
            y='stock_quantity',
            scatter_kws={'alpha': 0.5, 's': 20},
            line_kws={'color': 'red', 'linewidth': 2},
            ax=axes[1])
axes[1].set_title(f'Scatter plot: price vs stock_quantity\nКорреляция: {correlation_matrix.iloc[0,1]:.3f}', 
                    fontsize=14)
axes[1].set_xlabel('Цена (price)', fontsize=12)
axes[1].set_ylabel('Количество на складе (stock_quantity)', fontsize=12)
axes[1].grid(True, alpha=0.3)
    
plt.tight_layout()
plt.show()
    
# Интерпретация корреляции
print("\n🔍 ИНТЕРПРЕТАЦИЯ КОРРЕЛЯЦИИ:")
corr_value = correlation_matrix.iloc[0, 1]
    
if abs(corr_value) >= 0.7:
    strength = "сильная"
elif abs(corr_value) >= 0.3:
    strength = "умеренная"
elif abs(corr_value) >= 0.1:
    strength = "слабая"
else:
    strength = "очень слабая или отсутствует"
    
if corr_value > 0:
    direction = "положительная"
    interpretation = "чем выше цена, тем больше товаров на складе"
elif corr_value < 0:
    direction = "отрицательная"
    interpretation = "чем выше цена, тем меньше товаров на складе"
else:
    direction = "отсутствует"
    interpretation = "нет линейной связи между ценой и количеством на складе"
    
print(f"   Корреляция между price и stock_quantity: {corr_value:.3f}")
print(f"   Направление: {direction}")
print(f"   Сила связи: {strength}")
print(f"   Интерпретация: {interpretation}")
```

```python
# 2. ФИЛЬТРАЦИЯ ДАННЫХ ПО ЗАДАННОМУ ПЕРИОДУ
print("\n\n2. ФИЛЬТРАЦИЯ ДАННЫХ:")
print("-" * 40)

# Определяем период фильтрации
start_date = '2025-01-01'
end_date = '2025-06-01'  # Внимание: до 01 июня, не включая 01 июня

print(f"Период фильтрации: с {start_date} по {end_date} (не включая {end_date})")

# Преобразуем строки дат в datetime
try:
    start_dt = pd.to_datetime(start_date)
    end_dt = pd.to_datetime(end_date)
    
    print(f"Начальная дата (datetime): {start_dt}")
    print(f"Конечная дата (datetime): {end_dt}")
    
    # Фильтруем данные
    print("\nФильтрация заказов...")
    
    # Создаем маску для фильтрации
    mask = (orders['order_date'] >= start_dt) & (orders['order_date'] < end_dt)
    
    # Применяем фильтр
    filtered_orders = orders[mask].copy()
    
    print(f"✅ Фильтрация завершена")
    print(f"   Всего заказов до фильтрации: {len(orders)}")
    print(f"   Заказов за период: {len(filtered_orders)}")
    print(f"   Процент от общего числа: {len(filtered_orders)/len(orders)*100:.1f}%")
    
    # Проверяем, есть ли данные после фильтрации
    if len(filtered_orders) == 0:
        print("\n⚠️  ВНИМАНИЕ: После фильтрации не осталось ни одного заказа!")
        print("   Возможные причины:")
        print("   1. В данных нет заказов за указанный период")
        print("   2. Неправильный формат дат в исходных данных")
        print("   3. Диапазон дат в данных не пересекается с указанным периодом")
        
        # Показываем распределение заказов по годам и месяцам
        if orders['order_date'].notna().any():
            print("\n   Распределение заказов по годам:")
            orders_by_year = orders['order_date'].dt.year.value_counts().sort_index()
            for year, count in orders_by_year.items():
                print(f"     {year} год: {count} заказов")
            
            print("\n   Распределение заказов по месяцам (2025 год, если есть):")
            orders_2025 = orders[orders['order_date'].dt.year == 2025]
            if len(orders_2025) > 0:
                orders_by_month = orders_2025['order_date'].dt.month.value_counts().sort_index()
                month_names = ['Янв', 'Фев', 'Мар', 'Апр', 'Май', 'Июн', 'Июл', 'Авг', 'Сен', 'Окт', 'Ноя', 'Дек']
                for month, count in orders_by_month.items():
                    month_name = month_names[month-1] if month <= len(month_names) else f'Мес{month}'
                    print(f"     {month_name} 2025: {count} заказов")
    
except Exception as e:
    print(f"❌ Ошибка при фильтрации данных: {e}")
    filtered_orders = pd.DataFrame()  # Создаем пустой DataFrame
```

```python
# 3. СОХРАНЕНИЕ ДАННЫХ В CSV-ФАЙЛ
print("\n\n3. СОХРАНЕНИЕ ДАННЫХ В CSV-ФАЙЛ:")
print("-" * 40)

if len(filtered_orders) > 0:
    # Определяем имя файла
    filename = 'orders_2025_jan_to_may.csv'
    
    print(f"Сохранение данных в файл: {filename}")
    print(f"Формат: CSV с разделителем ';'")
    print(f"Кодировка: UTF-8")
    
    try:
        # Сохраняем данные в CSV с указанными параметрами
        filtered_orders.to_csv(
            filename,
            sep=';',           # Разделитель - точка с запятой
            index=False,       # Не сохранять индексы
            encoding='utf-8-sig'  # Кодировка с BOM для корректного отображения в Excel
        )
        
        print(f"✅ Файл успешно сохранен: {filename}")
        
        # Дополнительная информация о сохраненном файле
        import os
        if os.path.exists(filename):
            file_size = os.path.getsize(filename)
            file_size_mb = file_size / (1024 * 1024)  # Размер в MB
            
            print(f"\n📊 ИНФОРМАЦИЯ О СОХРАНЕННОМ ФАЙЛЕ:")
            print(f"   Имя файла: {filename}")
            print(f"   Размер файла: {file_size_mb:.2f} MB ({file_size} байт)")
            print(f"   Количество строк: {len(filtered_orders)}")
            print(f"   Количество столбцов: {len(filtered_orders.columns)}")
            print(f"   Столбцы в файле: {', '.join(filtered_orders.columns.tolist())}")
            
            # Проверяем первые несколько строк сохраненного файла
            print(f"\n🔍 ПРЕВЬЮ СОХРАНЕННЫХ ДАННЫХ (первые 3 строки):")
            print(filtered_orders.head(3).to_string())
            
            # Статистика по датам в сохраненных данных
            print(f"\n📅 СТАТИСТИКА ПО ДАТАМ В СОХРАНЕННЫХ ДАННЫХ:")
            if filtered_orders['order_date'].notna().any():
                saved_min_date = filtered_orders['order_date'].min()
                saved_max_date = filtered_orders['order_date'].max()
                print(f"   Самая ранняя дата: {saved_min_date}")
                print(f"   Самая поздняя дата: {saved_max_date}")
                print(f"   Период: {(saved_max_date - saved_min_date).days} дней")
                
                # Распределение по месяцам
                print(f"\n   Распределение заказов по месяцам:")
                months_dist = filtered_orders['order_date'].dt.to_period('M').value_counts().sort_index()
                for period, count in months_dist.items():
                    print(f"     {period}: {count} заказов")
            
            # Статистика по статусам заказов
            if 'status' in filtered_orders.columns:
                print(f"\n📊 СТАТИСТИКА ПО СТАТУСАМ ЗАКАЗОВ:")
                status_dist = filtered_orders['status'].value_counts()
                for status, count in status_dist.items():
                    percentage = (count / len(filtered_orders)) * 100
                    print(f"     {status}: {count} заказов ({percentage:.1f}%)")
            
            print(f"\n💾 Файл готов к использованию!")
            print(f"   Путь к файлу: {os.path.abspath(filename)}")
            
        else:
            print(f"❌ Ошибка: файл {filename} не был создан")
            
    except Exception as e:
        print(f"❌ Ошибка при сохранении файла: {e}")
        
else:
    print("❌ Нет данных для сохранения")
    print("   Создание пустого файла с заголовками...")
    
    try:
        # Создаем пустой файл с заголовками
        filename = 'orders_2025_jan_to_may_empty.csv'
        orders.head(0).to_csv(filename, sep=';', index=False, encoding='utf-8-sig')
        print(f"✅ Создан пустой файл с заголовками: {filename}")
    except Exception as e:
        print(f"❌ Ошибка при создании пустого файла: {e}")
```
